```python
#now we will do Pre-processing
#will remove GB from RAM , kg from Weight and remove the column of Unnamed
#here we are doing this to make object data-type to int data-type
```

```python
df.drop(columns=['Unnamed: 0'],inplace= True)
```

```python
#but here we can see our Ram and Weight is still object so first we have to change its data type
df['Ram']=df['Ram'].astype('int32')
df['Weight']=df['Weight'].astype('float32')
```

## Now will Do Feature Engg

```python
df['IPS panel']=df['ScreenResolution'].apply(lambda x:1 if 'IPS Panel' in x else 0)
```

```python
sns.barplot(x = df['IPS panel'], y =df['Price'])
```

```python
new = df['ScreenResolution'].str.split('x', n=1 , expand = True)
```

```python
df['X_res'] = new[0]
df['Y_res'] = new[1]
```

```python
df.head()
```

#### But now we also have to extract only numerical part

#### we are find the code basically

```python
df['X_res']=df['X_res'].str.replace(',', '').str.findall(r'(\d+\.?\d+)').apply(lambda x:x[0])
```

```python
df.head()
```

```python
df['X_res']=df['X_res'].astype('int')
df['Y_res']=df['Y_res'].astype('int')
```
